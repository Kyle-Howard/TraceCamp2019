```python
import random

def zodiac (m, d):
    if (m == 1):
        if (d < 20):
            sign = "Capricorn"
        else:
            sign = "Aquarius"
    elif (m == 2):
        if (d < 19):
            sign = "Aquarius"
        else: 
            sign = "Pisces"
    elif (m == 3):
        if (d < 21):
            sign = "Pisces"
        else: 
            sign = "Aries"
    elif (m == 4):
        if (d < 20):
            sign = "Aries"
        else: 
            sign = "Taurus"
    elif (m == 5):
        if (d < 21):
            sign = "Taurus"
        else: 
            sign = "Gemini"
    elif (m == 6):
        if (d < 21):
            sign = "Gemini"
        else: 
            sign = "Cancer"
    elif (m == 7):
        if (d < 23):
            sign = "Cancer"
        else: 
            sign = "Leo"
    elif (m == 8):
        if (d < 23):
            sign = "Leo"
        else: 
            sign = "Virgo"
    elif (m == 9):
        if (d < 23):
            sign = "Virgo"
        else: 
            sign = "Libra"
    elif (m ==10):
        if (d < 23):
            sign = "Libra"
        else: 
            sign = "Scorpio"
    elif (m == 11):
        if (d < 22):
            sign = "Scorpio"
        else: 
            sign = "Sagittarius"
    elif (m == 12):
        if (d < 22):
            sign = "Sagittarius"
        else: 
            sign = "Capricorn"
    else:
        sign = "N/A"
    
    return sign;

def horoscope(sign, messages):
    index = 2.99*random.random()
    index = int(index)
    print(index)
    print(f'As a {sign}, your horoscope today is: {messages[index]}\n')
    return 
    

scopes = ["You will be lucky today.", "You will be unlucky today.", "Today will be what you make it."]
month = input ('What month were you born in (as an integer, i.e. 1, 2, ..., 12)? ')
day = input ('What day were you born on? ')
try:
    month = int(month)
    day = int(day)
except ValueError:
    print("Invalid birthday!\n")

sign = zodiac(month, day)
horoscope(sign, scopes)


```
